```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Vanilla 32

```python
blue32 = pd.read_csv('Viz/Training/bluebert_vanilla_model32_version_3.csv') 
sci32 = pd.read_csv('Viz/Training/scibert_vanilla_model32_version_1.csv') 
```

```python
df = blue32
df2 = sci32

x = df['Step']
y = df['Value']
x2 = df2['Step']
y2 = df2['Value']
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training loss for vanilla variation 32')
plt.plot(x,y, label='BlueBERT')
plt.plot(x2,y2, label='SciBERT')
plt.legend()

plt.grid(color='#0f0f0f', linestyle='--', linewidth=0.15)
```

# Vanilla 16, 8, 4

```python
blue16 = pd.read_csv('Viz/Training/bluebert_vanilla_model16_version_1.csv') 
sci16 = pd.read_csv('Viz/Training/scibert_vanilla_model16_version_0.csv') 

blue8 = pd.read_csv('Viz/Training/bluebert_vanilla_model8_version_0.csv') 
sci8 = pd.read_csv('Viz/Training/scibert_vanilla_model8_version_0.csv') 

blue4 = pd.read_csv('Viz/Training/bluebert_vanilla_model4_version_0.csv') 
sci4 = pd.read_csv('Viz/Training/scibert_vanilla_model4_version_0.csv') 
```

```python
def plot(df1, df2, title, df3=0, df4=0, df5=0, df6=0):
    x = df1['Step']
    y = df1['Value']

    x2 = df2['Step']
    y2 = df2['Value']

    if not df3.empty:
        x3 = df3['Step']
        y3 = df3['Value']

        x4 = df4['Step']
        y4 = df4['Value']

        x5 = df5['Step']
        y5 = df5['Value']

        x6 = df6['Step']
        y6 = df6['Value']

    plt.xlabel('Step')
    plt.ylabel('Loss')
    plt.title(title)

    plt.plot(x,y, label='BlueBERT 16', color='C0')
    plt.plot(x2,y2, label='SciBERT 16', color='C1')

    if not df3.empty:
        plt.plot(x3,y3, label='BlueBERT 8', linestyle ='--', color='C0')
        plt.plot(x4,y4, label='SciBERT 8', linestyle ='--',color='C1')

        plt.plot(x5,y5, label='BlueBERT 4', linestyle =':', color='C0')
        plt.plot(x6,y6, label='SciBERT 4', linestyle =':',color='C1')

    plt.legend()

    plt.grid(color='#0f0f0f', linestyle='--', linewidth=0.15)
```

```python
df = blue16
df2 = sci16
plot(blue16, sci16, 'Training loss for vanilla variations 16, 8, & 4', blue8, sci8, blue4, sci4 )
```

# Regularised

```python
def plot(df1, df2, title, df1label, df2label, linestyle=None):
    x = df1['Step']
    y = df1['Value']

    x2 = df2['Step']
    y2 = df2['Value']


    plt.xlabel('Step')
    plt.ylabel('Loss')
    plt.title(title)

    plt.plot(x,y, label=df1label, color='C0', linestyle = linestyle)
    plt.plot(x2,y2, label=df2label, color='C1', linestyle = linestyle)

    plt.legend()

    plt.grid(color='#0f0f0f', linestyle='--', linewidth=0.15)
```

```python
blueReg = pd.read_csv('Viz/Training/bluebert_regularised_model_version_0.csv') 
sciReg = pd.read_csv('Viz/Training/scibert_regularised_model_version_0.csv') 
plot(blueReg, sciReg, 'Training loss with regularised variation', 'BlueBERT', 'SciBERT')
```

```python
blueRegVal = pd.read_csv('Viz/Validation/bluebert_regularised_model_version_0.csv') 
sciRegVal = pd.read_csv('Viz/Validation/scibert_regularised_model_version_0.csv') 
plot(blueRegVal, sciRegVal, 'Validation loss with regularised variation','BlueBERT', 'SciBERT')
```

# Masking, Oversampled, Undersampled

```python
maskedBlue = pd.read_csv('Viz/Training/bluebert_masked_model_version_0.csv')
maskedSci = pd.read_csv('Viz/Training/scibert_masked_model_version_1.csv')
maskedvalBlue = pd.read_csv('Viz/Validation/bluebert_masked_model_version_0.csv')
maskedvalSci = pd.read_csv('Viz/Validation/scibert_masked_model_version_1.csv')

overBlue = pd.read_csv('Viz/Training/bluebert_oversampled_model_val_version_0.csv')
overSci = pd.read_csv('Viz/Training/scibert_oversampled_model_val_version_0.csv')
overvalBlue = pd.read_csv('Viz/Validation/bluebert_oversampled_model_val_version_0.csv')
overvalSci = pd.read_csv('Viz/Validation/scibert_oversampled_model_val_version_0.csv')

underBlue = pd.read_csv('Viz/Training/bluebert_undersampled_model_val_version_0.csv')
underSci = pd.read_csv('Viz/Training/scibert_undersampled_model_val_version_0.csv')
undervalBlue = pd.read_csv('Viz/Validation/bluebert_undersampled_model_val_version_0.csv')
undervalSci = pd.read_csv('Viz/Validation/scibert_undersampled_model_val_version_0.csv') 
```

```python
plot(maskedBlue, maskedSci, 'Training loss of masked, oversampled, and undersampled data','masked BlueBERT', 'masked SciBERT')
plot(overBlue, overSci, 'Training loss of masked, oversampled, and undersampled data','Oversampled BlueBERT', 'Oversampled SciBERT','--')
plot(underBlue, underSci, 'Training loss of masked, oversampled, and undersampled data','Undersampled BlueBERT', 'Undersampled SciBERT',':')
```

```python
plot(maskedvalBlue,maskedvalSci, 'Validation loss of masked, oversampled, and undersampled data','masked BlueBERT', 'undersampled SciBERT' )
plot(overvalBlue,overvalSci, 'Validation loss of masked, oversampled, and undersampled data','undersampled BlueBERT', 'undersampled SciBERT','--' )
plot(undervalBlue,undervalSci, 'Validation loss of masked, oversampled, and undersampled data','undersampled BlueBERT', 'undersampled SciBERT',':' )
```
